```python
import pandas as pd 
import numpy as np
import matplotlib as plt
import warnings,os
warnings.filterwarnings('ignore')
data= pd.read_csv("..\\data\\final.csv")

data.head()

data.loc[320]['Call Drop Category']

data.shape

data_new = data[['Average Signal Strength Download','Average Signal Strength Upload','Average Data Speed Download','Average Data Speed Upload']]
from scipy import stats
data = data[(np.abs(stats.zscore(data_new)) < 3).all(axis=1)]

#data.to_csv('outlier.csv')



```

```python
Var_Corr = data.corr()

import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.heatmap(Var_Corr, xticklabels=Var_Corr.columns, yticklabels=Var_Corr.columns, annot=True)

data['Average_Signal'] = (data['Average Signal Strength Download']*0.955 +data['Average Signal Strength Upload']*0.045)/2

Var_Corr = data.corr()

import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.heatmap(Var_Corr, xticklabels=Var_Corr.columns, yticklabels=Var_Corr.columns, annot=True)

```

```python
sns.set_style('darkgrid')
sns.distplot(data['Average Signal Strength Upload'])

```

```python
sns.set_style('darkgrid')
sns.distplot(data['Average_Signal'])

```
